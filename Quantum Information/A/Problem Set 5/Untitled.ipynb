```python
import numpy as np

n = 100000
c = 2
count =0

#r is right hand side
#l is left hand side
for c in range(0,1000):
    l = c * (n**np.log(n))
    r = c**n 
    if l <= r:
        print('c = %d, YES'%(c))
        count +=1
    else:
        print('c = %d, NO'%(c))
print(count)
#Outputs YES for all values c > 1
```

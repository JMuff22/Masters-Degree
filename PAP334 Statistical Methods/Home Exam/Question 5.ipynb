```python
import numpy as np
import matplotlib.pyplot as plt

Dat1 = np.loadtxt("eccentricity.txt")
print(np.median(Dat1))
binwidth = 0.05

x = np.zeros(len(Dat1))
for i in range(len(Dat1)):
    x[i] = np.sqrt(Dat1[i])

    
plot = plt.hist(x, bins=np.arange(min(Dat1), max(Dat1) + binwidth, binwidth), histtype= 'step', color='black')
plt.ylabel('Counts')
plt.xlabel('x= $\sqrt{e}$')

c = max(plot[0])
d = plot[1][7]
#print(np.where(plot[0] == c))

print('c=%f' %d)
plt.vlines(plot[1][7], 0,70, colors='b')
```
